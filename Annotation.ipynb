```python
import openai
import pandas as pd
import time
import json

# ✅ Create a client with your GPT-4 API key
client = openai.OpenAI(api_key="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")  # <-- Replace with your actual API key

def extract_aspect_sentiment(review_text):
    prompt = f"""
Given the review below, extract the sentiment (Positive, Negative, Neutral, or None) for the following aspects:
Seat, Food, Staff/Crew Behavior, Check-in/Boarding, Cleanliness, In-flight Entertainment, Punctuality, Baggage Handling.

Review:
\"\"\"{review_text}\"\"\"

Output only valid JSON in this format:
{{
  "Seat": "...",
  "Food": "...",
  "Staff": "...",
  "Check-in": "...",
  "Cleanliness": "...",
  "Entertainment": "...",
  "Punctuality": "...",
  "Baggage": "..."
}}
"""

    try:
        response = client.chat.completions.create(
            model="gpt-4",
            messages=[{"role": "user", "content": prompt}],
            temperature=0
        )
        result = response.choices[0].message.content.strip()
        return json.loads(result)
    except Exception as e:
        print(f"Error at review: {e}")
        return {aspect: "None" for aspect in [
            "Seat", "Food", "Staff", "Check-in", "Cleanliness", "Entertainment", "Punctuality", "Baggage"
        ]}

# ✅ Load your dataset
df = pd.read_csv("/content/FINAL1.csv")
aspect_cols = ["Seat", "Food", "Staff", "Check-in", "Cleanliness", "Entertainment", "Punctuality", "Baggage"]

# ✅ Process each review
for idx, row in df.iterrows():
    sentiments = extract_aspect_sentiment(row["review_text"])
    for col in aspect_cols:
        df.at[idx, f"{col} Sentiment"] = sentiments.get(col, "None")
    print(f"✅ Processed review {idx + 1}/{len(df)}")
    time.sleep(1.2)

# ✅ Save the result
df.to_csv("annotated_reviews.csv", index=False)
print("✅ All reviews processed and saved.")
```
